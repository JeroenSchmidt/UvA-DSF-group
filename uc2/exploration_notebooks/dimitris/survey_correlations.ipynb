```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
%matplotlib inline
```

```python
survey = pd.read_pickle('../../data/Visual_well_being/survey.pickle')
```

```python
survey.info()
```

```python
survey_scores = survey[['P', 'E', 'R', 'M', 'A', 'PERMA', 'N_EMO', 'P_EMO']]
survey_scores = survey_scores[pd.notnull(survey_scores['PERMA'])]
```

## Correlations Exploration

```python
sns.heatmap(survey_scores.corr(), annot=True)
plt.show()
```

```python
sns.pairplot(survey_scores, kind='reg')
```

## More Instagram Images = Less Happiness ?

```python
plot = sns.jointplot(data=survey, x='imagecount', y='PERMA', kind='reg')
plot.annotate(stats.pearsonr)
plt.show()
```
