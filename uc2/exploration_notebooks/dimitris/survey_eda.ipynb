## Demographics Distribution

```python
plot = sns.countplot(x='income', data=survey, color = '#52839e', order = survey['income'].value_counts().index)
plot.set_title('Income Distribution')
plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
```

```python
plot = sns.countplot(x='education', data=survey, order = survey['education'].value_counts().index)
plot.set_title('Education Distribution')
plot.set_xticklabels(plot.get_xticklabels(), rotation=30)
```

```python
plot = sns.countplot(x='gender', data=survey, order = survey['gender'].value_counts().index)
plot.set_title('Gender Distribution')
```

```python
plot = sns.countplot(x='education', hue='gender', data=survey, order = survey['education'].value_counts().index)
plot.set_title('Education Distribution per Gender')
plot.set_xticklabels(plot.get_xticklabels(), rotation=30)
```

## PERMA Scores per Income

```python
survey['income'] = survey['income'].str.replace('I\'d rather not disclose this information', 'Undisclosed')

plot = sns.boxplot(data=survey, x='income', y='PERMA')
plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
plot.set_title('PERMA Scores per Income')
plt.show()
```

```python
survey['income'] = survey['income'].str.replace('I\'d rather not disclose this information', 'Undisclosed')

plot = sns.boxplot(data=survey, x='income', y='PERMA', 
                   order=['$150,000 or more', '$100,000 to $149,999', '$90,000 to $99,999', '$80,000 to $89,999',
                          '$70,000 to $79,999', '$60,000 to $69,999', '$50,000 to $59,999', '$40,000 to $49,999',
                          '$30,000 to $39,999', '$20,000 to $29,999', 'Less than $10,000'])
plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
plot.set_title('PERMA Scores per Income')
plt.show()
```

## PERMA Scores per Gender

```python
plot = sns.boxplot(data=survey, x='gender', y='PERMA')
plot.set_title('PERMA Scores per Gender')
plt.show()
```

## PERMA Scores per Employed 

```python
survey['employed'] = survey['employed'].str.replace('Out of work but not currently looking for work', 'Not looking for work')
survey['employed'] = survey['employed'].str.replace('Out of work and looking for work', 'Looking for work')

plot = sns.boxplot(data=survey, x='employed', y='PERMA')
plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
plot.set_title('PERMA Scores per Employed Status')
plt.show()
```

## PERMA Scores versus Education

```python
plot = sns.boxplot(data=survey, x='education', y='PERMA', 
                   order=['Post graduate degree', 'College graduate', 'High school graduate', 'Some high school'])
plot.set_xticklabels(plot.get_xticklabels(), rotation=45)
plot.set_title('PERMA Scores vs Education')
plt.show()
```
