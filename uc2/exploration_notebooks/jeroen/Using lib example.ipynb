# object percentages

There are over 2000 different topics/objects. The high dimensionality makes it impossible to explore easily w.r.t to other variables like quality of life. It is therfor best to pick specific topics and then compare them with what you want.

The following example looks at selfies

This will plot a histogram of the % that selfies consisted of peoples photos

```python
obj_p = usr_f.ratio_of_topics()
```

```python
ax = obj_p[["user_id","Selfie"]][obj_p.Selfie > 0].plot(kind="hist")
ax.set_xlabel("% of photos which are selfies")
ax.set_ylabel("number of users")
```
